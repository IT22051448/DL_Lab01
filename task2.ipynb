01. Create a new dataframe by filtering all the columns [i.e., features] except N_total, N_bulk,
N_surface and R_avg columns

```python
import pandas as pd
df = pd.read_csv('Au_nanoParticle_dataset.csv')
df.head()
```

```python
filtered_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
filtered_df.head()
```

```python
filtered_df.head(20)
```

03. Calculate the mean, standard deviation and quartile values for each of the above 4 features.


```python
filtered_df.describe().loc[['mean', 'std', '25%', '50%', '75%']]
```

04. Plot the histogram of each of these features in a 1x4 layout.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 5))

plt.subplot(1, 4, 1)
plt.hist(filtered_df['N_total'], bins=100, color='blue', alpha=0.7)
plt.title('N_total Distribution')
plt.xlabel('N_total')
plt.ylabel('Frequency')

plt.subplot(1, 4, 2)
plt.hist(filtered_df['N_bulk'], bins=100, color='red', alpha=0.7)
plt.title('N_bulk Distribution')
plt.xlabel('N_bulk')

plt.subplot(1, 4, 3)
plt.hist(filtered_df['N_surface'], bins=100, color='orange', alpha=0.7)
plt.title('N_surface Distribution')
plt.xlabel('N_surface')

plt.subplot(1, 4, 4)
plt.hist(filtered_df['R_avg'], bins=100, color='purple', alpha=0.7)
plt.title('R_avg Distribution')
plt.xlabel('R_avg')


plt.tight_layout()
plt.show()
```

05. Visualize the scatter plots and histograms of this dataframe using the pairplot functionality of seaborn library

```python
import seaborn as sns
sns.pairplot(filtered_df, diag_kind='hist', corner=False)

plt.suptitle("Pairplot of N_total, N_bulk, N_surface, R_avg", y=1.02, fontsize=14)
plt.show()
```

06. Add the below code and change it such that,


a. Plots on the diagonal contains the histogram of each feature along with the kernal
density estimation plot. <br>
b. Plots on the lower half [g.map_lower] contains the bivariate kernal density
estimation plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt


g = sns.PairGrid(filtered_df)

g.map_upper(sns.histplot)

g.map_diag(sns.histplot, kde=True)

g.map_lower(sns.kdeplot, fill=True, cmap="Blues", thresh=0.05)

plt.suptitle("PairGrid with Histograms, KDE, and Bivariate KDE", y=1.02, fontsize=14)

plt.show()
```
