02. Create a random 100000x1 array from the same exponential distribution and visualize the distribution histogram along with uniform and normal distributions. Use the code given below and do the necessary changes. Change the number of bins and the plot view to make the visualization better.

```python
import matplotlib.pyplot as plt

plt.hist(np.random.exponential(size=(100000, 1)), density=True, bins=50, histtype="step", color="blue",label="exponential")
plt.hist(np.random.randn(100000), density=True, bins=50, histtype="step", color="red",label="randn")
plt.hist(np.random.uniform(size=(100000,1)), density=True, bins=50, histtype="step", color="purple",label="uniform")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

03. Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5]

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

Z = X**2 + Y**2

figure = plt.figure(1, figsize=(10, 6))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1,
                                 cmap=matplotlib.cm.viridis, linewidth=0.1)


subplot3d.set_title("3D Plot of Z = X² + Y²")
subplot3d.set_xlabel("X-axis")
subplot3d.set_ylabel("Y-axis")
subplot3d.set_zlabel("Z-axis")

plt.show()
```

04. Calculate the pearson standard coefficient and spearman rank correlation for the [HP,
Attack, Defense, Sp. Ark, Sp. Def, Speed] features in the seaborn tutorial.ipynb. Visualize the
results using heatmaps. Make sure the values are displaced in the heatmap as well.


```python
import pandas as pd
# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
# Seaborn for plotting and styling
import seaborn as sns

# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Display first 5 observations
df.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = df[features]


pearson_corr = data.corr(method='pearson')
spearman_corr = data.corr(method='spearman')


plt.figure(figsize=(12, 5))


plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Pearson Correlation")


plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Spearman Correlation")

plt.tight_layout()
plt.show()
```
